```python
import pandas as pd
```

```python
df = pd.read_csv("World_development_dataset.csv")
df = df.dropna(subset=['Country Name'])
id_cols = ['Country Name', 'Country Code', 'Series Name', 'Series Code']
year_cols = [col for col in df.columns if ' [YR' in col]
df_long = df.melt(id_vars=id_cols,
                  value_vars=year_cols,
                  var_name='Year',
                  value_name='Value')
df_long['Year'] = df_long['Year'].str.split(' ').str[0]
df_long['Value'] = pd.to_numeric(df_long['Value'], errors='coerce')
df_final = df_long.pivot_table(index=['Country Name', 'Country Code', 'Year'],
                                columns='Series Name',
                                values='Value').reset_index()
df_final.columns.name = None
df_final
```

```python
df_final = df_final.assign(income = pd.qcut(df_final.get("GNI, Atlas method (current US$)"), q=5, labels=["very low", "low", "medium", "high", "very high"]))
df_final = df_final.assign(gdp_growth = pd.qcut(df_final.get('GDP growth (annual %)'), q=5, labels=["very low", "low", "medium", "high", "very high"]))
df_final = df_final.assign(inequality = pd.qcut(df_final.get('Income share held by lowest 20%'), q=5, labels=["very low", "low", "medium", "high", "very high"]))
df_final = df_final.assign(inflation = pd.qcut(df_final.get('Inflation, GDP deflator (annual %)'), q=5, labels= ["very high", "high", "medium", "low", "very low"]))
df_final = df_final.assign(life_exp = pd.qcut(df_final.get('Life expectancy at birth, total (years)'), q=5, labels=["very low", "low", "medium", "high", "very high"]))
df_final = df_final.assign(poverty = pd.qcut(df_final.get('Poverty headcount ratio at $3.00 a day (2021 PPP) (% of population)'), q=5, labels=["very high", "high", "medium", "low", "very low"]))
df_final = df_final.assign(school_enroll = pd.qcut(df_final.get('School enrollment, primary and secondary (gross), gender parity index (GPI)'), q=5, labels=["very low", "low", "medium", "high", "very high"]))
'''
def numeric(val):
  if val == "very low":
    return 0
  elif val == "low":
    return 1
  elif val == "medium":
    return 2
  elif val == "high":
    return 3
  elif val == "very high":
    return 4
df_final = df_final.assign(poverty_num = df_final.get("poverty").apply(numeric))
df_final = df_final.assign(school_enroll_num = df_final.get("school_enroll").apply(numeric))
df_final = df_final.assign(life_exp_num = df_final.get("life_exp").apply(numeric))
df_final = df_final.assign(inequality_num = df_final.get("inequality").apply(numeric))
df_final = df_final.assign(gdp_growth_num = df_final.get("gdp_growth").apply(numeric))
df_final = df_final.assign(inflation_num = df_final.get("inflation").apply(numeric))
df_final = df_final.assign(income_num = df_final.get("income").apply(numeric))
'''

CATEGORY_MAP = {
    "very low": 0,
    "low": 1,
    "medium": 2,
    "high": 3,
    "very high": 4
}

num_cols = [
    "income", "gdp_growth", "inequality",
    "inflation", "life_exp", "poverty", "school_enroll"
]

for col in num_cols:
    df_final[f"{col}_num"] = (
        df_final[col]
        .map(CATEGORY_MAP)
        .astype(float)
    )

df_final
sample = df_final.get([ "income_num",  "gdp_growth_num", "inequality_num",
    "inflation_num", "life_exp_num", "poverty_num", "school_enroll_num"])

df_final["income_num"] = df_final["income_num"].fillna(
    df_final["income_num"].median()
)

df_final

num_cols1 = [  "gdp_growth_num", "inequality_num",
    "inflation_num", "life_exp_num", "poverty_num", "school_enroll_num"]
for col in num_cols1:
  series = sample.groupby("income_num")[col].transform("median")
  df_final[col] = df_final[col].fillna(series)
df_final.to_csv("polished_dataset.csv")
```
